```python
#4번문제

import collections
counter = collections.Counter(title)
print(counter)
```

```python
webtoon_dict = {}
for counter in title:
    cur_webtoon_count = webtoon_dict.get(counter)
    
    if not cur_webtoon_count:
        webtoon_dict[counter] = 1
    else:
        webtoon_dict[counter] += 1

webtoon_dict
```

```python
import operator
sorted_dict = sorted(webtoon_dict.items(), reverse=True, key=operator.itemgetter(1))
```
